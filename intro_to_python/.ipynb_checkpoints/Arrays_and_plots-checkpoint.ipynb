Raw data from an instrument is often saved as a .txt or .csv file. Treating these 2-dimensional arrays of numbers as matrices allows you to perform all sorts of operations and filters on your data. This notebook will go use NumPy to load and analyze that data. Throughout these analyses, Matplotlib will be used to visualize the data

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

# %matplotlib inline allows the jupyter notebook's cell output generate the figure,
# rather than generating a new window for it
```

Let's make up some data, so that you can understand what it looks like as numpy acts on it

```python
x = np.arange(0, 10, 1) #create a 1D array from 0 to 9 with increments of 1
y = np.arange(0, 10, 1)

print (x)

fig, ax = plt.subplots() #create a figure and axes using plt.subplots()
plt.plot(x,y)            #to the figure, add a line plot of x vs y
plt.scatter(x,y)         #to the figure, add a scatter plot of x vs y
plt.show()               #show the final figure and axes
```

While you could for loops to access every element in the array, numpy uses vectorization to perform actions simultaneously on each element. This reduces the computational cost of the operation and save you time

```python
z1 = np.zeros(len(x))
z2 = np.zeros(len(x))
z3 = np.zeros(len(x))

for i in range(len(x)):
    z1[i] = x[i] + 2
    z2[i] = x[i] * 2
    z3[i] = x[i] ** 2
    
fig, ax = plt.subplots()
plt.plot(x, z1, c = 'r', label = 'x + 2') #note that you can specify the label or color of a data series using strings
plt.plot(x, z2, c = 'g', label = 'x * 2')
plt.plot(x, z3, c = 'b', label = '$x^2$') #matplotlib supports LaTeX math notation in label strings
plt.legend(loc = 'upper left')            #Legend showing the labels and colors for the data series in the figure
plt.show()
```

That for loop can be done more quickly and succintly using nd.array() functionality

```python
z4 = x + 2
z5 = x * 2
z6 = np.power(x,2)

fig, ax = plt.subplots()
plt.plot(x, z4, c = 'r', label = 'x + 2')
plt.plot(x, z5, c = 'g', label = 'x * 2')
plt.plot(x, z6, c = 'b', label = '$x^2$')
plt.legend(loc = 'upper left')
plt.show()
```

Numpy is great for holding datasets that contain only floats or ints. To support data with string values, you'll want to use pandas. When loading in tabular data, there is one main functions you can use: np.loadtxt(). This function will read in any 2-D you have. If the symbol that separates the column values, called the delimiter, is a tab, or a comma, or some other symbol, you can specify that to the loadtxt() function by using the flag "delimiter = " in the loadtxt function, followed by a string of the delimiter. Tab = '\t', comma = ' , ', space = ' ' and so on

```python
filepath = './JV_curve.liv1' #string filepath refers to this_directory/filename.
                             #Note that .liv1 is a raw file from the solar sim in the RTT. It is tab-delimited text

jv_curve = np.loadtxt(filepath, max_rows = 34, delimiter = '\t')
print (jv_curve)
```

In order to reference columns, rows, or values in a numpy array, you use what's called 'slice-indexing'. This is the same format as referencing elements in a python list or dictionary:

For example, x = np.zeros((10,10,10)) is a 10x10x10 matrix of zeroes. This is a 3-dimensional matrixs, meaning it has 3 axes. The very first element is called with x[0,0,0] or x[0][0][0]

```python
x = np.zeros((10,10,10))

print ('x.shape = ', x.shape)          #show the shape of the whole matrix
print ('x[0,0,0] = ', x[0,0,0])     #show the first element
print ('x[0][0][0] = ', x[0][0][0]) #also show the first element

print ('x[:,0,0] = ', x[:,0,0]) #show all of the elements in the first dimension, but only their first values in the second and third
print ('x[0:4, :, :] = ', x[0:4, :, :]) #show the first through the 5th element in the first dimension, and all of those values in the second and third

# in Slice-indexing, -1 refers to the final value in that axis, -2 the second to last
# Note that in a slice, _e.g._ x[-3:-1], all but the final value are shown. So, the second number in the slice, the stop index, is not included
print ('x[-1, -1, -1] = ', x[-1, -1, -3:-1])
```

Let's use slice-indexing to reference the x and y columns in our uploaded data

```python
x = jv_curve[:,0] #every row of the first column
y = jv_curve[:,1] #every row of the second column
y_flipped = y *- 1 #the data comes in "upside-down", so let's just flip it all


fig, ax = plt.subplots()
plt.scatter(x, y)
plt.plot(jv_curve[:,0], jv_curve[:,1]) #note that you can use slice indexing in your plotting functions!!
plt.title('Raw JV-curve')
plt.xlabel('Volts')
plt.ylabel('Current Density')
plt.show()

plt.plot(x, y_flipped)
plt.scatter(x, y_flipped)
plt.title('Flipped JV-curve')
plt.xlabel('Volts')
plt.ylabel('Current Density')
plt.show()
```

Now that we have the data loaded into Python, we need to perform some analyses on it. For OPV, that means calculating 4 values:
- $J_{sc}$ = J value at 0 V
- $V_{oc}$ = V value at 0 A
- FF = $\frac{J_{mp}*V_{mp}}{J_{sc}*V_{oc}}$, where mp is the maximum power point of the JV curve
- PCE = $J_{sc}*V_{oc}*FF$

But this is Python! We know we'll likely have to do this analysis a few times on different samples, so let's go ahead and just define a function for it. First we'll need some tools to make this easier on ourselves. The same group that maintains NumPy also maintains SciPy, which is scientific functions in python. If you need to do regressions or curve-fitting, SciPy is your best bet. (* for harder regressions and machine learning, look to scikit-learn!!)

I like to define functions after figuring out all of the parts within it. So let's walk through this process together

```python
from scipy.optimize import fsolve #function solver
from scipy.optimize import fmin #calculates the minimum for a function, given data
from scipy.interpolate import interp1d #interpolates 1D data
```

```python
# First, we need to find our J and V intercepts. We might not have any datapoints at those exact V = 0 or J = 0 values
# So we will need to interpolate between our datapoints to find that.

# first, add, or append, our flipped y data the dataset
print (jv_curve.shape)
print (y_flipped.shape)

data = np.c_[jv_curve, y_flipped] #add column. To add a row, use np.r_
print (data.shape)

#tell scipy.interp1d to create a line of best fit using a cubic function
JVinterp = interp1d(data[:,0], data[:,2], kind = 'cubic', bounds_error = False, fill_value = 'extrapolate')

#x = 0 (which is V=0 in our data) gives us our Jsc
Jsc = JVinterp(0)
print (f"Jsc = {Jsc}") #this type of formatting is called "f-strings". they interpret how to print the variable in {}

#y = 0 (which is J=0 in our data) gives us our Voc. Use scipy.fsolve to identify this point
Voc = fsolve(JVinterp,.95*max(data[:,0]))
print (f"Voc = {Voc}")

zeros = np.zeros(len(data[:,0])) #we'll need this for visualizint our axes and origin

fig, ax = plt.subplots(figsize = (8,6))
plt.plot(data[:,0], data[:,2]) #plot the data
plt.plot(zeros, data[:,2], c = 'k') #plot x-axis
plt.plot(data[:,0], zeros, c = 'k') #plot y-axis
plt.scatter(0, Jsc, c = 'r') #plot the Jsc
plt.scatter(Voc, 0, c = 'r') #plot the Voc
plt.title('JV-curve with $J_{sc}$ and $V_{oc}$')
plt.xlabel('Volts')
plt.ylabel('Current Density')
plt.show()
```

```python
# Now let's calculate our FF and PCE. To do that, we need to find the point of maximum power

#minimize the function that calculates power output at each datapoint
PPV = fmin(lambda x: x*JVinterp(x), .8*Voc, disp=False) #lambda functions are small, hidden functions that don't need to be called elsewhere
print (PPV)

Vmax = PPV #x-coordinate of maximum power point
Jmax = JVinterp(PPV) #y-coordinate of maximum power point

# PCE is the maximum theoretical power * the actual maximum power. minus sign cancels out negative current values
PCE = -Jmax*Vmax
print (PCE)

# FF is the PCE divided by the observed max power (== Jsc*Voc), and is a percentage. minus sign cancel out negative current values
FF = (-Jmax*Vmax)/(Jsc*Voc)*100
print (FF)
```

```python
zeros = np.zeros(len(data[:,0])) #we'll need this for visualizint our axes and origin

fig, ax = plt.subplots(figsize = (8,6))
plt.plot(data[:,0], data[:,2]) #plot the data
plt.plot(zeros, data[:,2], c = 'k') #plot x-axis
plt.plot(data[:,0], zeros, c = 'k') #plot y-axis
plt.scatter(0, Jsc, c = 'r') #plot the Jsc
plt.scatter(Voc, 0, c = 'r') #plot the Voc
plt.scatter(Vmax, Jmax, c = 'r') #plot the maximum power point
plt.title('JV-curve with $J_{sc}$ and $V_{oc}$')
plt.xlabel('Volts')
plt.ylabel('Current Density')
plt.show()
```

# So, now that we have code to evaluate our data and calculate our device metrics, let's compile that all into a single function

To do that, I'm literally just copying and pasting those lines of code beneath my function definition line. I'm also going to describe the function, usage, and input/output variables in a doc-string. A doc-string is a block of documentation text that is surrounded by 2 sets of 3 quotationmarks (referred to as 'triple-doubles'). This just denotes a block of string-text.

```python
def calculate_jv_values(jv_curve):
    """
    This is a function that takes in an np.array of JV-curve data and calculates Jsc, Voc, FF, and PCE.
    
    Parameters
    ----------
    jv_curve : np.ndarray()
        A 2-column array, where column 1 is voltage values and column 2 is current values
        
    Returns
    -------
    metrics : list of floats
        A 4-element list containg the PCE, Voc, Jsc, and FF values for the input J-V curve. 
    """
    y = jv_curve[:,1]
    y_flipped = y *- 1
    
    data = np.c_[jv_curve, y_flipped]

    JVinterp = interp1d(data[:,0], data[:,2], kind = 'cubic', bounds_error = False, fill_value = 'extrapolate')
    Jsc = JVinterp(0)
    Voc = fsolve(JVinterp,.95*max(data[:,0]))
    PPV = fmin(lambda x: x*JVinterp(x), .8*Voc, disp=False)
    Vmax = PPV
    Jmax = JVinterp(PPV)
    PCE = -Jmax*Vmax
    FF = (Jmax*Vmax)/(Jsc*Voc)*100

    metrics = [float(PCE), float(Voc), float(Jsc), float(FF)] #converts array-like variables to a list of floats
        
    return metrics
```

So, now we can just call all that code in one line!

```python
test = np.loadtxt(filepath, max_rows = 34, delimiter = '\t')

metrics = calculate_jv_values(test)

print (metrics)
```
