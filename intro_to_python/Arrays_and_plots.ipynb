Raw data from an instrument is often saved as a .txt or .csv file. Treating these 2-dimensional arrays of numbers as matrices allows you to perform all sorts of operations and filters on your data. This notebook will go use NumPy to load and analyze that data. Throughout these analyses, Matplotlib will be used to visualize the data

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

# %matplotlib inline allows the jupyter notebook's cell output generate the figure,
# rather than generating a new window for it
```

Let's make up some data, so that you can understand what it looks like as numpy acts on it

```python
x = np.arange(0, 10, 1) #create a 1D array from 0 to 9 with increments of 1
y = np.arange(0, 10, 1)

print (x)

fig, ax = plt.subplots() #create a figure and axes using plt.subplots()
plt.plot(x,y)            #to the figure, add a line plot of x vs y
plt.scatter(x,y)         #to the figure, add a scatter plot of x vs y
plt.show()               #show the final figure and axes
```

While you could for loops to access every element in the array, numpy uses vectorization to perform actions simultaneously on each element. This reduces the computational cost of the operation and save you time

```python
z1 = np.zeros(len(x))
z2 = np.zeros(len(x))
z3 = np.zeros(len(x))

for i in range(len(x)):
    z1[i] = x[i] + 2
    z2[i] = x[i] * 2
    z3[i] = x[i] ** 2
    
fig, ax = plt.subplots()
plt.plot(x, z1, c = 'r', label = 'x + 2') #note that you can specify the label or color of a data series using strings
plt.plot(x, z2, c = 'g', label = 'x * 2')
plt.plot(x, z3, c = 'b', label = '$x^2$') #matplotlib supports LaTeX math notation in label strings
plt.legend(loc = 'upper left')            #Legend showing the labels and colors for the data series in the figure
plt.show()
```

That for loop can be done more quickly and succintly using nd.array() functionality

```python
z4 = x + 2
z5 = x * 2
z6 = np.power(x,2)

fig, ax = plt.subplots()
plt.plot(x, z4, c = 'r', label = 'x + 2')
plt.plot(x, z5, c = 'g', label = 'x * 2')
plt.plot(x, z6, c = 'b', label = '$x^2$')
plt.legend(loc = 'upper left')
plt.show()
```

Numpy is great for holding datasets that contain only floats or ints. To support data with string values, you'll want to use pandas. When loading in tabular data, there is one main functions you can use: np.loadtxt(). This function will read in any 2-D you have. If the symbol that separates the column values, called the delimiter, is a tab, or a comma, or some other symbol, you can specify that to the loadtxt() function by using the flag "delimiter = " in the loadtxt function, followed by a string of the delimiter. Tab = '\t', comma = ' , ', space = ' ' and so on

```python
filepath = './JV_curve.liv1'

jv_curve = np.loadtxt(filepath, max_rows = 34, delimiter = '\t')
print (jv_curve)
```

In order to reference columns, rows, or values in a numpy array, you use what's called 'slice-indexing'. This is the same format as referencing elements in a python list or dictionary:

For example, x = np.zeros((10,10,10)) is a 10x10x10 matrix of zeroes. This is a 3-dimensional matrixs, meaning it has 3 axes. The very first element is called with x[0,0,0] or x[0][0][0]

```python
x = np.zeros((10,10,10))

print ('x.shape = ', x.shape)          #show the shape of the whole matrix
print ('x[0,0,0] = ', x[0,0,0])     #show the first element
print ('x[0][0][0] = ', x[0][0][0]) #also show the first element

print ('x[:,0,0] = ', x[:,0,0]) #show all of the elements in the first dimension, but only their first values in the second and third
print ('x[0:4, :, :] = ', x[0:4, :, :]) #show the first through the 5th element in the first dimension, and all of those values in the second and third

# in Slice-indexing, -1 refers to the final value in that axis, -2 the second to last
# Note that in a slice, _e.g._ x[-3:-1], all but the final value are shown. So, the second number in the slice, the stop index, is not included
print ('x[-1, -1, -1] = ', x[-1, -1, -3:-1])
```

Let's use slice-indexing to reference the x and y columns in our uploaded data

```python
x = jv_curve[:,0] #every row of the first column
y = jv_curve[:,1] #every row of the second column
y_flipped = y *- 1


fig, ax = plt.subplots()
plt.scatter(x, y)
plt.plot(jv_curve[:,0], jv_curve[:,1]) #note that you can use slice indexing in your plotting functions!!
plt.show()

plt.plot(x, y_flipped)
plt.scatter(x, y_flipped)
plt.show()
```
