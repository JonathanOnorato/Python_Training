Raw data from an instrument is often saved as a .txt or .csv file. Treating these 2-dimensional arrays of numbers as matrices allows you to perform all sorts of operations and filters on your data. This notebook will go use NumPy to load and analyze that data. Throughout these analyses, Matplotlib will be used to visualize the data

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

# %matplotlib inline allows the jupyter notebook's cell output generate the figure,
# rather than generating a new window for it
```

Let's make up some data, so that you can understand what it looks like as numpy acts on it

```python
x = np.arange(0, 10, 1) #create a 1D array from 0 to 9 with increments of 1
y = np.arange(0, 10, 1)

print (x)

fig, ax = plt.subplots()
plt.plot(x,y)
plt.show()

fig, ax = plt.subplots()
plt.scatter(x,y)
plt.show()
```

While you could for loops to access every element in the array, numpy uses vectorization to perform actions simultaneously on each element. This reduces the computational cost of the operation and save you time

```python
z1 = np.zeros(len(x))
z2 = np.zeros(len(x))
z3 = np.zeros(len(x))

for i in range(len(x)):
    z1[i] = x[i] + 2
    z2[i] = x[i] * 2
    z3[i] = x[i] ** 2
    
fig, ax = plt.subplots()
plt.plot(x, z1, c = 'r', label = 'x + 2')
plt.plot(x, z2, c = 'g', label = 'x * 2')
plt.plot(x, z3, c = 'b', label = '$x^2$')
plt.legend(loc = 'upper left')
plt.show()
```

That for loop can be done more quickly and succintly using nd.array() functionality

```python
z4 = x + 2
z5 = x * 2
z6 = np.power(x,2)

fig, ax = plt.subplots()
plt.plot(x, z4, c = 'r', label = 'x + 2')
plt.plot(x, z5, c = 'g', label = 'x * 2')
plt.plot(x, z6, c = 'b', label = '$x^2$')
plt.legend(loc = 'upper left')
plt.show()
```
